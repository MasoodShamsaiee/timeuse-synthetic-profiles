# Time Use Harmonization

Convert the raw 2022 TUS PUMF files into harmonized respondent and episode tables.

```python
%load_ext autoreload
%autoreload 2

from pathlib import Path
import sys

CWD = Path.cwd().resolve()
PROJECT_ROOT = CWD.parent if CWD.name.lower() == "notebooks" else CWD
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd

from src.timeuse_profiles.harmonize import (
    harmonize_tus2022_episode_pumf,
    harmonize_tus2022_main_pumf,
    save_harmonized_tables,
)
from src.timeuse_profiles.io import read_table
```

```python
RAW_DIR = PROJECT_ROOT / "data" / "raw" / "time_use"
OUT_DIR = RAW_DIR
RAW_RESPONDENTS_PATH = RAW_DIR / "Data_Données" / "TU_ET_2022_Main-Principal_PUMF.sas7bdat"
RAW_EPISODES_PATH = RAW_DIR / "Data_Données" / "TU_ET_2022_Episode_PUMF.sas7bdat"
```

```python
raw_resp = read_table(RAW_RESPONDENTS_PATH)
raw_epi = read_table(RAW_EPISODES_PATH)

harm_resp = harmonize_tus2022_main_pumf(raw_resp)
harm_epi = harmonize_tus2022_episode_pumf(raw_epi)
resp_path, epi_path = save_harmonized_tables(harm_resp, harm_epi, out_dir=OUT_DIR)

display(harm_resp.head())
display(harm_epi.head())
print(resp_path)
print(epi_path)
```
